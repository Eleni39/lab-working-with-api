```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# finding the url to the source that we need

url = 'http://www.popvortex.com/music/charts/top-100-songs.php'
```

```python
response = requests.get(url)
response.status_code 
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
#checking the soup
soup.prettify()
```

```python
soup.select("#chart-position-2 > div.chart-content.col-xs-12.col-sm-8 > p > cite.title")
```

```python
soup.select("cite.title")
```

```python
soup.select("em.artist")
```

```python
title = []
artist = []

num_iter = len(soup.select("cite.title"))
    
for i in range(num_iter):
    title.append(soup.select("cite.title")[i].get_text())
    artist.append(soup.select("em.artist")[i].get_text())
    
print(title)
print(artist)
    
```

```python
top100 = pd.DataFrame({"title":title,
                           "artist":artist,
                      })
```
